```python
import math

def has_nan(lst):
    for x in lst:
        if isinstance(x, float) and math.isnan(x):
            return True
    return False


if has_nan(unique_list):
    print("The list contains NaN values.")
else:
    print("The list does not contain NaN values.")
```

```python
def remove_nan(lst):
    return [x for x in lst if not (isinstance(x, float) and math.isnan(x))]

cleaned_list = remove_nan(unique_list)
```
